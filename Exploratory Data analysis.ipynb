```python
#map
d = {'no': False, 'yes': True}
df['international plan'] = df['international plan'].map(d)
df.head()
```

```python
#Daatframe Transformation
total_calls = df['total day calls'] + df['total eve calls'] + \
df['total night calls'] + df['total intl calls']
print(total_calls)
df.insert(loc = len(df.columns), column = 'total_calls', value = total_calls)
```

```python
df.head()
```

```python
#drop
df.drop('total_calls', axis = 1, inplace = True) #axis = 0 deletes index and axis = 1 deletes columns
# inplace = False (default)
# inplace = True (do the operation and replace none)
```
